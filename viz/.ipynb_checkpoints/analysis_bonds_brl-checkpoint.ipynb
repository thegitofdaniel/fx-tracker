# MACRO VARIABLES

```python
df = pd.read_parquet("../data/silver/macro_brl.parquet")

df["real_selic"]=(1+df["interest_brl"])/(1+df["inflation_yoy_brl"])-1
print(df.columns)

plot_timeseries(df[["interest_brl","inflation_yoy_brl"]]
                ,title="SELIC, Inflation, and Real (MoM, %)"
                ,plot_diff=True
                ,scale_factor=100)

plot_timeseries(pd.DataFrame(df["real_selic"])
                ,title="Returns of SELIC Bonds (Real, Annualized, %)"
                ,scale_factor=100
                ,plot_avg_year=["real_selic"])

plot_histogram(df["real_selic"]
               ,title = "Interest Rate Real Returns (Annualized)")
```

# PRE

```python
df_bonds = pd.read_parquet("../data/silver/bonds_brl.parquet")

# add useful variables
#df_bonds["time_to_maturity"]=(df_bonds["maturity_date"]-df_bonds["date"]).dt.days
df_bonds["bond_name"] = df_bonds["maturity_date"].dt.strftime('%Y_%m_%d')+"_"+df_bonds.bond_type

df_bonds.tail()
```

```python
df_pre = df_bonds[df_bonds["bond_type"]=="pre"]
df_pre = df_pre.groupby("date").min()
df_pre = df_pre[["bid_yield","inflation_brl","real_return"]]
df_pre.head()

df_pre_annual = (1+df_pre)**12-1

plot_timeseries(df_pre_annual
                ,title="Macro Variables- Monthly Values"
                ,scale_factor=100)

plot_histogram(df_pre_annual["real_return"]
               ,title="Returns of PRE Bonds (Real, Annualized, %)")

plot_timeseries(pd.DataFrame(df_pre_annual["real_return"])
                ,title="Macro Variables - Monthly Values"
                ,scale_factor=100,
                plot_avg_year=["real_return"])
```

```python
df_pre = df_bonds[df_bonds["bond_type"]=="pre"]
df_pre = df_pre.groupby("date").mean()
df_pre = df_pre[["bid_yield","inflation_brl","real_return"]]
df_pre.head()

df_pre_annual = (1+df_pre)**12-1

plot_timeseries(df_pre_annual
                ,title="Macro Variables- Monthly Values"
                ,scale_factor=100)

plot_histogram(df_pre_annual["real_return"]
               ,title="Returns of PRE Bonds (Real, Annualized, %)")

plot_timeseries(pd.DataFrame(df_pre_annual["real_return"])
                ,title="Macro Variables - Monthly Values"
                ,scale_factor=100,
                plot_avg_year=["real_return"])
```
