```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import MonthLocator, YearLocator, DateFormatter
from pandas.tseries.offsets import MonthBegin
import numpy as np
```

```python
dom_cur="usd"
for_cur="brl"

df = pd.read_parquet("../data/gold/macro_"+dom_cur+for_cur+".parquet")
df.tail()
```

```python
# df
piv = pd.pivot_table(df, index=df.index.year,columns=df.index.month, values=["fxreal_"+dom_cur+for_cur])
piv.columns = piv.columns.droplevel()

# add DECEMBER of previous year (i.e. YEAR-END) for STANDARDIZATION
piv[0]=piv[12].shift(1)
piv = piv[piv[0].notnull()]
piv = piv.divide(piv[0],axis=0)
piv = piv[sorted(piv.columns)]

#########################################################

# hardcoded variables
fontsize=12
figsize=(24,12)

# create figure
fig = plt.figure(figsize=figsize)
ax = plt.axes()

# plot
piv.transpose().plot(ax=ax)

# graph elements
plt.title("Real FX Evolution Throughout the Year, Standardized",fontsize=fontsize+6)
plt.xlim(1, 12)
plt.xticks(np.arange(0, 12, 1.0))
ax.grid(True)

plt.show()
```

```python
# observerd probability of real FX INCREASE in a given month
# INCREASE -> STRONGER DOMESTIC CURRENCY (i.e. weak BRL)
for i in range(1,13):
    print("{:>02} | {:.2f}%".format(i,(((piv[i]-piv[i-1])>0).sum()/piv.shape[0]*100)))
```

```python
# observerd probability of real FX DECREASE in a given month
# DECREASE -> WEAKER DOMESTIC CURRENCY (i.e. strong BRL)
for i in range(1,13):
    print("{:>02} | {:.2f}%".format(i,(((piv[i]-piv[i-1])<0).sum()/piv.shape[0]*100)))
```
